```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.figure(figsize=(8,5))
emp_id=np.arange(1,10)
salary=np.arange(20,110,10)
plt.bar(emp_id,salary)
plt.xlabel('Employee_ID')
plt.ylabel('Employee_Salary')
plt.show()
```

```python
plt.figure(figsize=(8,5))
ax=plt.axes()
ax.set_facecolor('grey')
plt.bar(emp_id,salary,color='red')
plt.xlabel('Employee_ID')
plt.ylabel('Employee_Salary')
plt.show()
```

```python
x1=([1,3,5,7])
x2=([2,4,6,8])
y1=([7,8,7,8])
y2=([18,17,16,15])
plt.figure(figsize=(8,6))
ax=plt.axes()
plt.bar(x1,y1,label='First',color='blue')
plt.bar(x2,y2,label='Second',color='orange')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Bar chart')
plt.legend()
plt.show()
```
