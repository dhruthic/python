```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
y=pd.Series([1,2,3,4,5,3])
x=pd.Series([1,2,3,4,5,6])
correlation=y.corr(x)
correlation
```

```python
plt.scatter(x,y)
plt.plot(np.unique(x),np.poly1d(np.polyfit(x,y,1))
        (np.unique(x)),color='green')
```

```python
plt.scatter(x,y)
plt.plot(np.unique(x),np.poly1d(np.polyfit(x,y,1))
        (np.unique(x)),color='yellow')
plt.xlabel('x axis')
plt.ylabel('y axis')
```
